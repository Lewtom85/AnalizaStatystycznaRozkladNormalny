Rozklad normalny o sredniej 3 i odchyleniu standardowy 1 dla 100 oraz 10 000.

Biblioteki:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Parametry rozk≈Çadu normalnego

```python
srednia = 3
odchylenie_standardowe = 1
```

Rozklad prob

```python
rozmiar_prob_1 = 100
rozmiar_prob_2 = 10000
```

Generowanie danych z rozkladu normalnego

```python
prob_1 = np.random.normal(srednia, odchylenie_standardowe, rozmiar_prob_1)
prob_2 = np.random.normal(srednia, odchylenie_standardowe, rozmiar_prob_2)
```

Tworzenie histogramow.

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(prob_1, bins=20, density=True, color='skyblue', edgecolor='black')
plt.title(f'Rozk≈Çad normalny (n={rozmiar_prob_1})')
plt.xlabel('Warto≈õƒá')
plt.ylabel('Gƒôsto≈õƒá prawdopodobie≈Ñstwa')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.hist(prob_2, bins=50, density=True, color='lightcoral', edgecolor='black')
plt.title(f'Rozk≈Çad normalny (n={rozmiar_prob_2})')
plt.xlabel('Warto≈õƒá')
plt.ylabel('Gƒôsto≈õƒá prawdopodobie≈Ñstwa')
plt.grid(True)
```

Tworzenie wykresow pudelkowych.

```python
plt.subplot(2, 2, 3)
plt.boxplot(prob_1, vert=False)
plt.title(f'Wykres pude≈Çkowy (n={rozmiar_prob_1})')
plt.xlabel('Warto≈õƒá')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.boxplot(prob_2, vert=False)
plt.title(f'Wykres pude≈Çkowy (n={rozmiar_prob_2})')
plt.xlabel('Warto≈õƒá')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Podsumowanie, wnioski:
Im wieksza proba tym rozklad bardziej przypomina rozklad normalny.
